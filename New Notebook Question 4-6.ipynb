```python
%%sql

with cte as (
select distinct m.customerid, c.category, 
count(c.category) over (partition by m.customerid) as count
from messages m
join customers c 
on m.customerid = c.id
and category  like 'new'
order by count desc)

select avg(count)
from cte
```
